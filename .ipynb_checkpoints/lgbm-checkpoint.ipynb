```python
def preproc(dataFrame):
    saveIdx = dataFrame.index
    dataFrame = dataFrame.reset_index(drop=True)
    # 파생변수 생성
    dataFrame['u-g'] = dataFrame['u'] - dataFrame['g']
    dataFrame['g-r'] = dataFrame['g'] - dataFrame['r']
    dataFrame['r-i'] = dataFrame['r'] - dataFrame['i']
    dataFrame['i-z'] = dataFrame['i'] - dataFrame['z']
    # dataFrame['delta'] = dataFrame['nDetect'] - dataFrame['nObserve']
    # if dropD:
    #   dataFrame = dataFrame.drop(['dered_u','dered_g','dered_r','dered_i','dered_z'], axis=1)

    # PCA
    pca = PCA(n_components=1)
    df_pca = pca.fit_transform(dataFrame[['u', 'g', 'r', 'i', 'z']])
    df_pca2 = pca.fit_transform(dataFrame[['dered_u','dered_g','dered_r','dered_i','dered_z']])
    df_pca3 = pca.fit_transform(dataFrame[['airmass_u', 'airmass_g', 'airmass_r', 'airmass_i', 'airmass_z']])


    dataFrame = pd.concat((dataFrame, pd.DataFrame(df_pca)), axis = 1)
    dataFrame.rename({0:'F1'}, axis = 1, inplace = True)
    dataFrame = pd.concat((dataFrame, pd.DataFrame(df_pca2)), axis = 1)
    dataFrame.rename({0:'D1'}, axis = 1, inplace = True)
    dataFrame = pd.concat((dataFrame, pd.DataFrame(df_pca3)), axis = 1)
    dataFrame.rename({0:'A1'}, axis = 1, inplace = True)

    dataFrame.drop(['u', 'g', 'r', 'i', 'z', 'dered_u','dered_g','dered_r','dered_i','dered_z'], axis = 1, inplace = True)
    dataFrame.drop(['airmass_u','airmass_g','airmass_r','airmass_i','airmass_z'], axis = 1, inplace = True)

    dataFrame.index = saveIdx
    return dataFrame

# train = preproc(train)
# train
```
